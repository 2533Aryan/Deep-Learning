**a. Give a definition for probability:**

Probability is a measure of the likelihood that a specific event will occur. It is expressed as a value between 0 and 1, where 0 means the event cannot happen, and 1 means the event is certain to happen. Probability is commonly used to quantify uncertainty in various real-world contexts, such as games, experiments, or natural phenomena.

**b. Give a definition for sample spaces:**

A sample space is the set of all possible outcomes of a random experiment. For example, if we roll a six-sided die, the sample space consists of the numbers {1, 2, 3, 4, 5, 6}, which represent all the outcomes that could occur.

**c. Give a definition for events:**

An event is a subset of the sample space and consists of one or more outcomes. For instance, in the case of rolling a die, the event "rolling an even number" includes the outcomes {2, 4, 6}. Events can range from simple (one outcome) to complex (multiple outcomes grouped together).

**d. Give a definition for randomness:**

Randomness refers to the unpredictability or lack of pattern in the occurrence of events. In probability, randomness ensures that each outcome in the sample space has a fair and unbiased chance of occurring. For example, flipping a fair coin produces random results because the outcomes "heads" and "tails" are equally likely.

**3. Simulating Coin Tosses:** Demonstrate a simple probability experiment â€“ coin tossing. Simulate coin tosses in Python and calculate probabilities of getting Heads or Tails. Then, visualize the results by making a bar chart.

Write three different simulations (and three different bar charts):

a. 500 tosses

b. 1000 tosses

c. 2000 tosses.

## 500 tosses

```python
def simulate_coin_tosses(num_tosses):
    # Simulate coin tosses
    outcomes = np.random.choice(['Heads', 'Tails'], size=num_tosses)
    
    # Calculate probabilities
    head_prob = np.sum(outcomes == 'Heads') / num_tosses
    tail_prob = np.sum(outcomes == 'Tails') / num_tosses
    
    # Create a bar chart
    plt.bar(['Heads', 'Tails'], [head_prob, tail_prob], color=['blue', 'orange'])
    plt.title(f'Coin Toss Simulation ({num_tosses} Tosses)')
    plt.xlabel('Outcome')
    plt.ylabel('Probability')
    plt.ylim(0, 1)
    plt.show()

# Simulations for 500, 1000, and 2000 tosses
simulate_coin_tosses(500)
simulate_coin_tosses(1000)
simulate_coin_tosses(2000)
```

**4. Dice Rolling Simulation:** Extend the experiment to simulate rolling a six-sided die. Calculate the probabilities of rolling each number and visualize the results using bar charts. Again, do this for three scenarios:

a. 500 rolls

b. 1000 rolls

c. 2000 rolls

https://www.geeksforgeeks.org/bar-plot-in-matplotlib/


```python
# Simulates rolling a six-sided die and visualizes the probabilities.

def simulate_die_rolls(num_rolls):
    # Simulate die rolls
    roll_results = np.random.choice([1, 2, 3, 4, 5, 6], size=num_rolls)
    
    # Calculate probabilities
    roll_probs = [np.mean(roll_results == i) for i in range(1, 7)]
    
    # Create a bar chart
    plt.bar(range(1, 7), roll_probs, color=['red', 'green', 'blue', 'orange', 'purple', 'yellow'])
    plt.title(f'Dice Roll Simulation ({num_rolls} Rolls)')
    plt.xlabel('Die Face')
    plt.ylabel('Probability')
    plt.ylim(0, 1)
    plt.show()

# Run the function for different scenarios
simulate_die_rolls(500)  # Simulate 500 rolls
simulate_die_rolls(1000)  # Simulate 1000 rolls
simulate_die_rolls(2000)  # Simulate 2000 rolls
```

**a. Give a definition for probability distributions:**

A probability distribution is a mathematical function that describes how the probabilities of different outcomes are distributed for a random variable. It assigns a probability to each possible value (or range of values) of the random variable, ensuring that the total probability adds up to 1. For example, a probability distribution can show the likelihood of rolling each face on a six-sided die.

**b. Give a definition for uniform distribution:**

A uniform distribution is a type of probability distribution where all outcomes are equally likely. In a discrete uniform distribution, such as rolling a fair six-sided die, each outcome (1 through 6) has the same probability. For a continuous uniform distribution, all values within a specific range have an equal probability, and values outside that range have a probability of 0.

**c. Give a definition for normal distribution:**

A normal distribution, also known as a Gaussian distribution, is a probability distribution that is symmetric and bell-shaped, centered around a mean (average). In this distribution, most values cluster around the mean, and the likelihood of extreme values decreases as you move further away from the center. It is characterized by two parameters: the mean and the standard deviation, which determine its center and spread, respectively. Normal distributions are widely used in statistics and natural phenomena, such as measuring heights or test scores.

**7. Simulating a Normal Distribution**

Simulate data from a normal distribution. Calculate the mean and standard deviation of the simulated data (notice what happens!). Visualize the data in a histogram. Do this for three different sample sizes:

a. 500 samples

b. 1000 samples

c. 2000 samples

```python
# Simulate data from a normal distribution
# There are three hyperparameters to define when generating a normal distribution:
# mean, standard deviation, and the number of samples.

# Use np.random.normal() to generate the data.


def simulate_normal_distribution(num_samples, mean_param=0, std_param=1):
    # Simulate data
    data = np.random.normal(loc=mean_param, scale=std_param, size=num_samples)
    
    # Calculate descriptive statistics
    mean = np.mean(data)
    std_dev = np.std(data)
    
    # Print calculated statistics
    print(f"Sample Size: {num_samples}")
    print(f"Calculated Mean: {mean:.2f}")
    print(f"Calculated Standard Deviation: {std_dev:.2f}")
    
    # Create a histogram
    plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7, 
             label=f'Mean: {mean:.2f}, Std Dev: {std_dev:.2f}')
    plt.title(f'Normal Distribution Simulation ({num_samples} Samples)')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Run the function for different sample sizes
simulate_normal_distribution(500)
simulate_normal_distribution(1000)
simulate_normal_distribution(2000)
```
